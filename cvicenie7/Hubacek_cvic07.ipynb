```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread('rampa.png', 0)
img2 = cv.imread('predmety.png', 1)
img2_gray = cv.imread('predmety.png', 0)
```

```python
# _, img2 = cv.threshold(img1, threshold, maxval, cv.THRESH_BINARY)  # _ podciarovnik je pomocna premenna
prahy = [100, 100, 100, 100]

_, img_prah1 = cv.threshold(img1, prahy[0], 255, cv.THRESH_BINARY_INV)
_, img_prah2 = cv.threshold(img1, prahy[1], 255, cv.THRESH_TRUNC)
_, img_prah3 = cv.threshold(img1, prahy[2], 255, cv.THRESH_TOZERO)
_, img_prah4 = cv.threshold(img1, prahy[3], 255, cv.THRESH_TOZERO_INV)

plt.subplot(2, 2, 1)
plt.title("Prah = %i" % prahy[0] + "\n(cv.THRESH_BINARY_INV)", fontsize=10)
plt.tight_layout()
plt.imshow(img_prah1, "gray")

plt.subplot(2, 2, 2)
plt.title("Prah = %i" % prahy[1] + "\n(cv.THRESH_TRUNC)", fontsize=10)
plt.tight_layout()
plt.imshow(img_prah2, "gray")

plt.subplot(2, 2, 3)
plt.title("Prah = %i" % prahy[2] + "\n(cv.THRESH_TOZERO)", fontsize=10)
plt.tight_layout()
plt.imshow(img_prah3, "gray")

plt.subplot(2, 2, 4)
plt.title("Prah = %i" % prahy[3] + "\n(cv.THRESH_TOZERO_INV)", fontsize=10)
plt.tight_layout()
plt.imshow(img_prah4, "gray")

print('')
```

```python
#_, img3 = cv.threshold(img1, 50, 255, cv.THRESH_BINARY)  # do pomocnej premennej sa ulozi hodnota prahu
_, img4 = cv.threshold(img2_gray, 80, 255, cv.THRESH_BINARY)
_, img6 = cv.threshold(img2_gray, 170, 255, cv.THRESH_BINARY)
prah, img5 = cv.threshold(img2_gray, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)  # v tomto pripade sa do premennej prah ulozi automaticky vypocitany prah

print("rucny prah =%i" % _)
print("automaticky prah =%i" % prah)

# cv.imshow("Rampa", img1)
# cv.imshow("Predmety", img2)
# cv.imshow("Rampa - Binary", img3)
# cv.imshow("Predmety - Binary", img4)
# cv.imshow("Predmety - OTSU", img5)

plt.subplot(2, 2, 1)
plt.title("Predmety povodny", fontsize=10)
plt.tight_layout()
plt.imshow(img2_gray, "gray")

plt.subplot(2, 2, 2)
plt.title("Predmety prahovane rucne, prah = 80", fontsize=10)
plt.tight_layout()
plt.imshow(img4 , "gray")

plt.subplot(2, 2, 3)
plt.title("Predmety sive prahovane rucne, prah = 170", fontsize=10)
plt.tight_layout()
plt.imshow(img6, "gray")

plt.subplot(2, 2, 4)
plt.title("Predmety - OTSU= %i" % prah, fontsize=10)
plt.tight_layout()
plt.imshow(img5, "gray")


plt.savefig("prahovanie.jpg", dpi=250)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
kernelE4 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (4, 4))
kernelE10 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (10, 10))
kernelE25 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (25, 25))

print(kernelE4)
print(kernel4)
print(kernelE10)
print(kernel10)
print(kernelE25)
print(kernel25)


img9 = cv.imread("circles.jpg", 0)

img10 = cv.morphologyEx(img9, cv.MORPH_OPEN, kernelE4)
img11 = cv.morphologyEx(img9, cv.MORPH_CLOSE, kernelE4)
img12 = cv.morphologyEx(img9, cv.MORPH_OPEN, kernelE10)
img13 = cv.morphologyEx(img9, cv.MORPH_CLOSE, kernelE10)
img14 = cv.morphologyEx(img9, cv.MORPH_OPEN, kernelE25)
img15 = cv.morphologyEx(img9, cv.MORPH_CLOSE, kernelE25)

cv.imshow("Circles", img9)

cv.imshow("o EL.(4,4)", img10)
cv.imshow("c EL.(4,4)", img11)
cv.imshow("o EL.(10,10)", img12)
cv.imshow("c EL.(10,10)", img13)
cv.imshow("o EL.(25,25)", img14)
cv.imshow("c EL.(25,25)", img15)

cv.waitKey(0)
cv.destroyAllWindows()
```
